ML Model

SVM

```python
from sklearn.metrics import(accuracy_score,classification_report,confusion_matrix,precision_score,f1_score,recall_score,roc_auc_score,auc)
```

```python
importantFeatures = ['Air Pollution','Dust Allergy','Shortness of Breath','Wheezing','Frequent Cold','Dry Cough']
AirDataNew = AirData[importantFeatures]
AirDataNew.head()
```

```python
from sklearn.model_selection import train_test_split
y = AirData[['chronic Lung Disease']]
x_train,x_test,y_train,y_test = train_test_split(AirDataNew,y,test_size = 0.25)
```

```python
from sklearn.svm import LinearSVC
SVC = LinearSVC()
SVC.fit(x_train,y_train)
y_pred = SVC.predict(x_test)
```

```python
SVCAccuracy = accuracy_score(y_test,y_pred)
print("Accuracy : ",accuracy_score(y_test,y_pred),"\n")
print("Classification Report : \n",classification_report(y_test,y_pred))
print("Confusion Matrix : \n",confusion_matrix(y_test,y_pred),"\n")
print("Precision Score : ",precision_score(y_test,y_pred,average = None),"\n")
print("Recall Score : ",recall_score(y_test,y_pred,average = None),"\n")
print("F1 Score : ",f1_score(y_test,y_pred,average = None),"\n")
```

Decision Tree ML

```python
from sklearn.tree import DecisionTreeClassifier
DC = DecisionTreeClassifier()
DC.fit(x_train,y_train)
y_pred = DC.predict(x_test)
print
DCAccuracy = accuracy_score(y_test,y_pred)
print("Accuracy : ",accuracy_score(y_test,y_pred),"\n")
print("Classification Report : \n",classification_report(y_test,y_pred))
print("Confusion Matrix : \n",confusion_matrix(y_test,y_pred),"\n")
print("Precision Score : ",precision_score(y_test,y_pred,average = None),"\n")
print("Recall Score : ",recall_score(y_test,y_pred,average = None),"\n")
print("F1 Score : ",f1_score(y_test,y_pred,average = None),"\n")
```

Random Forest ML

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train,y_train)
y_pred_rf = RF.predict(x_test)

RFAccuracy = accuracy_score(y_test,y_pred_rf)
print("Accuracy : ",accuracy_score(y_test,y_pred_rf),"\n")
print("Classification Report : \n",classification_report(y_test,y_pred_rf))
print("Confusion Matrix : \n",confusion_matrix(y_test,y_pred_rf),"\n")
print("Precision Score : ",precision_score(y_test,y_pred_rf,average = None),"\n")
print("Recall Score : ",recall_score(y_test,y_pred_rf,average = None),"\n")
print("F1 Score : ",f1_score(y_test,y_pred_rf,average = None),"\n")
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
regg = LogisticRegression(random_state = 4)
regg.fit(x_train,y_train)
y_pred_LR = regg.predict(x_test)
print('Accuracy: {}'.format(regg.score(x_test, y_test)))
print(classification_report(y_test, y_pred_LR))
```
